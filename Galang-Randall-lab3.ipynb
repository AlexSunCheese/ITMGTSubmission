```python
legs = {
    ("upd", "admu"): 
    {
        "traveltimemins": 10
    },
    ("admu", "dlsu"): 
    {
        "traveltimemins": 35
    },
    ("dlsu", "upd"): 
    {
        "traveltimemins": 55
    },
}

def eta(source, destination):
    mins = 0
    s = False
    d = False
    
    for k, v in legs.items():
        (cs, cd) = k
        if cs == source:
            mins += v["traveltimemins"]
            source = cd
            s = True
        if cd == destination and s:
            d = True
            break
    if not d:
            for k, v in legs.items():
                (cs, cd) = k
                if cs == source:
                    mins += v["traveltimemins"]
                if cd == destination:
                    d = True
                    break
    return mins

source = input("Enter Source: ")
destination = input("Enter Destination: ")
print("ETA: ", eta(source, destination))
```
